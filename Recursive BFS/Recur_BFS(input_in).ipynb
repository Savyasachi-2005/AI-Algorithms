```python
def RBFS(start, goal_test, suc, heur):
    def rbfs_recur(node, g, f_limit, path):
        f = g + heur(node)
        print(f"Visiting {node}, g={g}, h={heur(node)}, f={f}, limit={f_limit}")
        if f > f_limit:
            return f, None
        if goal_test(node):
            return 'FOUND', path
        suc_list = []
        for ch, cost in suc(node):
            if ch not in path:
                suc_list.append((ch, g + cost, heur(ch)))
        if not suc_list:
            return float('inf'), None
        suc_list.sort(key=lambda x: x[1] + x[2])
        while True:
            best_suc, best_g, best_h = suc_list[0]
            best_f = best_g + best_h
            second_best_f = suc_list[1][1] + suc_list[1][2] if len(suc_list) > 1 else float('inf')
            result, result_path = rbfs_recur(best_suc, best_g, min(f_limit, second_best_f), path + [best_suc])
            if result == 'FOUND':
                return 'FOUND', result_path
            suc_list[0] = (best_suc, best_g, result)
            suc_list.sort(key=lambda x: x[1] + x[2])
            if suc_list[0][1] + suc_list[0][2] > f_limit:
                return suc_list[0][1] + suc_list[0][2], None

    result, path = rbfs_recur(start, 0, float('inf'), [start])
    if result == 'FOUND':
        return path
    return None

# Inbuilt graph and heuristic input
def get_hardcoded_input():
    start_node = 'A'
    goal_node = 'G'

    # Graph definition: node -> list of (child, cost)
    graph = {
        'A': [('B', 1), ('C', 4)],
        'B': [('D', 2), ('E', 5)],
        'C': [('F', 1)],
        'D': [('G', 7)],
        'E': [],
        'F': [('G', 3)],
        'G': []
    }

    # Heuristic values
    heuristics = {
        'A': 7,
        'B': 6,
        'C': 5,
        'D': 4,
        'E': 7,
        'F': 3,
        'G': 0
    }

    def goal_test(node):
        return node == goal_node

    def successors(node):
        return graph.get(node, [])

    def heuristic(node):
        return heuristics.get(node, float('inf'))s

    return start_node, goal_test, successors, heuristic

# Run RBFS with hardcoded input
start_node, goal_test, successors, heuristic = get_hardcoded_input()
result = RBFS(start_node, goal_test, successors, heuristic)

if result:
    print("\nGoal found! Path:", " -> ".join(result))
else:
    print("\nNo solution found.")
```
